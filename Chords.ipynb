### Imports

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import (
    Audio, display, clear_output)
from ipywidgets import widgets
import time
from functools import partial
%matplotlib inline
```

### Define constants

```python
rate = 16000.
duration = .5
arpeggio_pause = .5
t = np.linspace(
    0., duration, int(rate * duration))
```

## Function to play a note of frequency 'f' hertz 

```python
def synth(f):
    x = np.sin(f * 2. * np.pi * t)
    display(Audio(x, rate=rate, autoplay=True))
    print('This note\'s frequency is: ' + str(f))
```

## Functions to define the three triads

```python
#define chords: major, minor, diminished, 7th, M7, m7
def major_chord(f):
    synth(f)
    time.sleep(arpeggio_pause)
    synth(f*5/4)
    time.sleep(arpeggio_pause)
    synth(f*3/2)
    time.sleep(arpeggio_pause)

def minor_chord(f):
    synth(f)
    time.sleep(arpeggio_pause)
    synth(f*192/160)
    time.sleep(arpeggio_pause)
    synth(f*3/2)
    time.sleep(arpeggio_pause)
    
def dim_chord(f):
    synth(f)
    time.sleep(arpeggio_pause)
    synth(f*6/5)
    time.sleep(arpeggio_pause)
    synth(f*231/160)
    time.sleep(arpeggio_pause)
```

## 7th Chords
These are shortcuts to create seventh chords

```python
def dom_7(f):
    major_chord(f)
    synth(f*9/5)
    time.sleep(arpeggio_pause)
    
def min_7(f):
    dim_chord(f)
    synth(f*9/5)
    time.sleep(arpeggio_pause)

def maj_7(f):
    major_chord(f)
    synth(f*15/8)
    time.sleep(arpeggio_pause)
```

### Adding intervals
* dim9, dom9, aug9
* dim11, dom11, aug11
* dim13, dom13, aug13

**Chord name skeleton:**
* Letter (A)
* Chord flavor (major, minor)
* Interval stack (7th, 9th, 11th, 13th)
* Stack flavor (m, M, A)

*A7 = A, min, M, 7th,


### Advanced_chords

```python
def advanced_chords(f, flavor, interval_7 = None, interval_9 = None, interval_11 = None, interval_13 = None):
    if flavor == 'maj' or 'major':
        major_chord(f)
    elif flavor == 'min' or 'minor':
        minor_chord(f)
    elif flavor == 'dim' or 'diminished':
        dim_chord(f)
    else:
        print('you must have done something wrong')
        return

#Sevenths    
    if interval_7 == 'm':
        synth(f*16/9)
        time.sleep(arpeggio_pause)
    elif interval_7 == 'M':
        synth(f*9/5)
        time.sleep(arpeggio_pause)
    elif interval_7 == 'A':
        synth(f*531441/262144)
        time.sleep(arpeggio_pause)
    else:
        print('end of chord')
        
#Ninths
    if interval_9 == 'm':
        synth(f*256/243*2)
        time.sleep(arpeggio_pause)
    elif interval_9 == 'M':
        synth(f*9/4)
        time.sleep(arpeggio_pause)
    elif interval_9 == 'A':
        synth(f*19683/16384*2)
        time.sleep(arpeggio_pause)
    else:
        print('end of chord')

#Elevenths
    if interval_11 == 'm':
        synth(f*8192/6561*2)
        time.sleep(arpeggio_pause)
    elif interval_11 == 'M':
        synth(f*8/3)
        time.sleep(arpeggio_pause)
    elif interval_11 == 'A':           #tritone
        synth(f*729/512*2)
        time.sleep(arpeggio_pause)
    else:
        print('end of chord')

#Thirteenths
    if interval_13 == 'm':
        synth(f*128/81*2)
        time.sleep(arpeggio_pause)
    elif interval_13 == 'M':
        synth(f*27/16*2)
        time.sleep(arpeggio_pause)
    elif interval_13 == 'A':
        synth(f*59049/32768*2)
        time.sleep(arpeggio_pause)
    else:
        print('end of chord')
        return
```
